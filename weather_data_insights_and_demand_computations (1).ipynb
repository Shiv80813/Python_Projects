```python
#importing os
import os
#defining path
precipitation_folder = 'total_precipitation'
#creating a list of temporory storage
list = []
#using for loop to access one file at a time
for filename in os.listdir(precipitation_folder):
    #joining the file name to the path
    filename = os.path.join(precipitation_folder, filename)
    #reading the individual file
    df = pd.read_csv(filename, delimiter=';')  
    # Rename Month columns from German to English
    df = df.rename(columns={'Jahr': 'Year',
                            'Jan': 'Jan',
                            'Feb': 'Feb',
                            'Mrz': 'Mar',
                            'Apr': 'Apr',
                            'Mai': 'May',
                            'Jun': 'Jun',
                            'Jul': 'Jul',
                            'Aug': 'Aug',
                            'Sep': 'Sep',
                            'Okt': 'Oct',
                            'Nov': 'Nov',
                            'Dez': 'Dec',
                           })
    #appending the list
    list.append(df)
#concating the data for making a list
tp_data = pd.concat(list, ignore_index=True)
#displaying the head
tp_data.head()
```

```python
#importing os
import os
#defining path
vapor_folder = 'vapor_pressure'
#creating a list of temporory storage
list = []
#using for loop to access one file at a time
for filename in os.listdir(vapor_folder):
    #joining the file name to the path
    filename = os.path.join(vapor_folder, filename)
    #reading the individual file
    df = pd.read_csv(filename, delimiter=';')  
    # Rename Month columns from German to English
    df = df.rename(columns={'Jahr': 'Year',
                            'Jan': 'Jan',
                            'Feb': 'Feb',
                            'Mrz': 'Mar',
                            'Apr': 'Apr',
                            'Mai': 'May',
                            'Jun': 'Jun',
                            'Jul': 'Jul',
                            'Aug': 'Aug',
                            'Sep': 'Sep',
                            'Okt': 'Oct',
                            'Nov': 'Nov',
                            'Dez': 'Dec',
                           })
    #appending the list
    list.append(df)
#concating the data for making a list
vp_data = pd.concat(list, ignore_index=True)
#displaying the head
vp_data.head()
```

Next, we created a new dataframe that combines the three dataframes into one dataframe that meets the following specifications for ease of use:


* The dataframe has the following columns
    - Year
    - Month
    - Total_Precipitation
    - Vapor_Pressure
    - Sunshine_Duration
* The dataframe has data only for those years for which we have files for all three quantities.  

```python
#Checking the column names
vp_data.columns
```

```python
#Unpivoting the data as per requirement
pivot_vp_data=vp_data.melt(id_vars='Year', value_vars=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'], var_name='Month', value_name='Vapor Pressure')

pivot_sd_data=sd_data.melt(id_vars='Year', value_vars=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'], var_name='Month', value_name='Sunshine Duration')

pivot_tp_data=tp_data.melt(id_vars='Year', value_vars=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'], var_name='Month', value_name='Total Precipitation')

merged_data = pd.merge(pivot_tp_data, pivot_sd_data, on=['Year', 'Month'], how='inner')

# Then, merging the result into one dataframe
combined_data = pd.merge(merged_data, pivot_vp_data, on=['Year', 'Month'], how='inner')
combined_data
```

Next we will handle missing values in the new dataframe using linear interpolation. 

Linear interpolation makes sense here because there may be a time trend in the data. So we want to fill missing values with values that would follow the trend. You should use the pandas `DataFrame.interpolate()` function for dataframes. Specify 'linear' as the `method` and 'both' as the `limit_direction` to perform the linear interpolation. 

```python
#Converting the month to date time format to ensure the sorting of the data
combined_data['Month1'] = pd.to_datetime(combined_data['Month'], format='%b')
#Sorting the data as per dates for interpolation to work
combined_data=  combined_data.sort_values(by =['Year','Month1'])
combined_data
```

```python
#interpolating the data as per the requirement
interpolated_data = combined_data[['Year','Month','Total Precipitation', 'Sunshine Duration', 'Vapor Pressure']].interpolate(method = 'linear',limit_direction = 'both')
interpolated_data
```

## Computations

We have been provided an demand forecasting model that incorporates the climate data into their sales. Their analysis shows that there was a significant shift in customer behavior in the year 2000. As a result, they created two demand forecasting models - one that applies to years before 2000 and one that applies to year 2000 and after.

**Before 2000:** $ \text{forecasted demand} = 1000 - 10.5 * \text{total precipitation} + 1.009 * \text{sunshine duration} $


**2000 and after:** $ \text{forecasted demand} = 1000 - 10.5 * \text{total precipitation} + 1.19 * \text{sunshine duration} $

```python
#Using the formula after filtering the data as per given in question
Before2000 = 1000 - 10.5 * interpolated_data[interpolated_data['Year'] < 2000]['Total Precipitation'] + 1.009 * interpolated_data[interpolated_data['Year'] < 2000]['Sunshine Duration']
#In cases where the demand is less than 0, updating them to 0, because demand cannot be negative
Before2000[Before2000<0] = 0

#Using the formula after filtering the data as per given in question
After2000 = 1000 - 10.5 * interpolated_data[interpolated_data['Year'] >= 2000]['Total Precipitation'] + 1.19 * interpolated_data[interpolated_data['Year'] >= 2000]['Sunshine Duration']
#In cases where the demand is less than 0, updating them to 0, because demand cannot be negative
After2000[After2000<0] = 0
```

```python
#Joining the two to create one column
interpolated_data['Demand_Forecast'] = pd.concat([Before2000, After2000])
interpolated_data
```

Lets deal with demand uncertainity now. Typically future demand for a product is uncertain. The uncertainty of demand is captured by the standard deviation. The forecasting models discussed above provide an estimate of the standard deviation as well. Create a new column called StdDev_Demand and assign standard deviation values as follows. The standard deviation of demand for all years before 2000 should be 20. The standard deviation of demand for years 2000 or later should be 30.

```python
#Calculating the std dev as per the given formula and conditions and saving the data in a new column
interpolated_data['StdDev_Demand'] = np.where(interpolated_data['Year'] < 2000, 20, 30)
interpolated_data
```

Next, when the demand for a product is uncertain, the stocking quantity (quantity to produce or buy from a supplier) depends on the proability distribution of demand. 
For normally distributed demand, the quantity can be calculated using the mean and standard deviation of demand and a safety stock multiplier. The demand forecast and standard deviations values that you generated earlier can be used as estimates of the mean and standard deviation of demand for a given month and year. 

Assuming a safety stock multiplier of 1.65, we can calculate the stocking quantity using the following formula. 

$\text{Stocking Quantity} = \text{Mean Demand} + 1.65 * \text{Std Dev of Demand} $


```python
#Calculating the Stocking qty as per the given formula and conditions and saving the data in a new column
interpolated_data['Stocking_Quantity'] = interpolated_data['Demand_Forecast'] + (1.65 * interpolated_data['StdDev_Demand'])
interpolated_data
```
